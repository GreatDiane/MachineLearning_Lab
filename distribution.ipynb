```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
distro = np.random.binomial(1, .5, 10000)
plt.hist(distro, 2)
```

```python
plt.figure()
distro = np.random.binomial(4, .5, 10000)
plt.hist(distro, 5)
plt.show()
```

```python
plt.figure()
distro = np.random.multinomial(n=2, pvals=[1/6,1/6,1/6,1/6,1/6,1/6], size = 10000)
plt.hist(distro, 3)
plt.show()
```

```python
import random
plt.figure()
uniform_low=0.25
uniform_high=0.8
uniform = np.random.uniform(uniform_low, uniform_high,10000) 
plt.hist(uniform, 20)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
mu=0.
sigma=2.
distro = np.random.normal(mu, sigma, 10000)
plt.hist(distro, 100)
plt.show()
```
